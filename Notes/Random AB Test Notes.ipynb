# A/B Testing

1. Create two versions of one piece of content (control and experimental group) that are differentiates by one **independent variable**.
2. Identify your goal
3. State your test (z, t, chi-square)
4. Split your sample groups equally and randomly
5. Choose your sample size (big enough to gather statistically significant results) or let your test run long enough to achieve that sample size.
6. Decide how significant your results need to be
    - Confidence level of 95% minimum which means $\alpha$ = 0.05
    - Will want higher confidence rate for something that only slightly improves a metric
7. During A/B test
    - Test variations at the same time. Ex: Time of day is the same if tested 2 months apart

- e-commerce conversion rates
- lead conversion rates
- bounce rates

# Hypothesis Testing

1. Null Hypothesis
    - $\mu$ =  x if One-Tailed Test
    - $\mu$ >= x if Two-Sided Test
2. Alternative Hypothesis
    - $\mu$ != x if One-Tailed Test
    - $\mu$ <  x if Two-Sided Test
3. Choose significance level: $\alpha$ = 0.05
4. Analyze sample
    - t-stat
5. Find p value
    - Estimate using t-table
    - OR use python method (shown in t-test)
6. Compare $\alpha$ to p value
    - If p < $\alpha$, reject the null hypothesis
    - See example in t-test image

Debrief
- Two types of Errors
    1. **Type I Error**: Reject the Null Hypothesis (False Positive)
    2. **Type II Error**: Accept the Null Hypothesis (False Negative)
- **Statistical Power**: Probability that a test will correctly reject a false null hypothesis (aka Pr(True Positive)
    - Usually want this to be at least 80%
    - Statistical Power = 1 - Type II Error
    - Pr(True Positive) = 1 - Pr(False Negative)
   
| Actually      | Accept Null (Negative) | Reject Null (Positive)    |
| :---        |    :----              |          :---             |
| True        | True Negative          | True Positive <br />(**Statistical Power**)   |
| False       | False Negative <br />**Pr(Type II Error)**     | False Positive <br />**Pr(Type I Error)**     |
    
- Power Analysis
    - Use 3 of the following to calculate 1 of the others:
        1. Effect Size
        2. Sample Size
        3. Significance
        4. Statistical Power
    
    
# When to use what:

### Z-test
- sample size is large (n>50), population variance is known

### t-test
- sample size is small (n<50), population variance is unknown




### Power Analysis
Use Desired statistical significance, Statistical Power, and Effect size to determine Sample Size.

Takeaways
- "The takeaway here is that the more radical the change, the less scientific we need to be process-wise. The more specific the change (button color, microcopy, etc.), the more scientific we should be because the change is less likely to have a large and noticeable impact on conversion rate."
